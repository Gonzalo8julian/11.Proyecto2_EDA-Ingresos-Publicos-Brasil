# Nulos

```python
def reporte(df):
    df_report = pd.DataFrame()
    df_report["numero_nulos"] = df.isnull().sum()
    df_report["porcentaje_nulos"]= round((df.isnull().sum()/df.shape[0])*100,2)
    df_report["tipo_variables"] = pd.DataFrame(df.dtypes)
    
    return df_report
```

```python
reporte(df)
```
