- Análisis de las categorías de ingresos más significativas y su participación en los ingresos totales.

```python
categorias_ingresos = df.groupby("CATEGORÍA ECONÓMICA")[["VALOR PREVISTO ACTUALIZADO","VALOR LANZADO","VALOR REALIZADO"]].sum()
categorias_ingresos["DIFERENCIA"] = categorias_ingresos["VALOR PREVISTO ACTUALIZADO"] - categorias_ingresos["VALOR REALIZADO"]
categorias_ingresos = categorias_ingresos.reset_index()
categorias_ingresos.sort_values(by="DIFERENCIA", ascending=False)
```

Vemos que en todos los casos se ha superado el valor lanzado, excepto en aquellas categorías sin información, que se estimó una cantidad que no se ha alcanzado por muchísima diferencia.

En el caso de las Receitas Correntes - intra-orçamentárias y de las Receitas de Capital - intra-orçamentárias  el valor realizado ha superado incluso al valor previsto actualizado, en este segundo caso pasando de 10,808,411,537.00 a 28,250,835,724.33. 

```python
origen_ingresos = df.groupby("ORIGEN INGRESO")[["VALOR PREVISTO ACTUALIZADO","VALOR LANZADO","VALOR REALIZADO"]].sum()
origen_ingresos = origen_ingresos.reset_index()

origen_ingresos["DIFERENCIA"] = origen_ingresos["VALOR PREVISTO ACTUALIZADO"] - origen_ingresos["VALOR REALIZADO"]
origen_ingresos.sort_values(by="DIFERENCIA", ascending=False)
```

En el caso del análisis del origen de los ingresos podemos ver una conclusión similiar a la anterior. Vemos que en la mayoría de los casos se ha superado el valor lanzado. Pero hay tres orígenes que la diferencia entre el valor previsto actualizado y el valor realizado es negativo, lo cuál quiere decir que se ha superado el valor previsto.

```python
df["MES"] = df["FECHA LANZAMIENTO"].dt.month_name()
df["MES_NUMERO"] = df["FECHA LANZAMIENTO"].dt.month
df.sample(1)
```
